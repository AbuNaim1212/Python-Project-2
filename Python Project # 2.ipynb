```python
#Import Tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Import the data file 
df=pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv', index_col=False)
```

```python
#Sorting by Gender
#determine gender types
df.groupby(['gender'])['gender'].count().sort_values(ascending = False)
```

```python
#Sorting by Race
df.groupby(['race'])['race'].count().sort_values(ascending = False)
```

```python
#Data Cleaning# Lebel '?' as 'unknown'
df.loc[df['race'] == '?', 'race'] = 'unknown'
df.groupby(['race'])['race'].count().sort_values(ascending = False)
```

```python
#Creating new data set
df1 = df[['race', 'gender', 'number_diagnoses']]
print(df1)
```

```python
# Data set by Gender
df2=df1[['gender', 'number_diagnoses']]
print(df2)
```

```python
#Visualization by Gender
df2.groupby(df['gender']).count().plot(kind='bar', title = 'Chart-1')
```

```python
# Data set by Race
df3=df1[['race', 'number_diagnoses']]
print(df3)
```

```python
#Visualization by Race
df3.groupby(df['race']).count().plot(kind='bar', title = 'Chart-2')
```

```python
#Findings: According to 'Chat-1' Female gender has higher 'Diabetes' diagnoses number and 'Char-2' Caucasian race has higher 'Diabetes' diagnoses number.
```
